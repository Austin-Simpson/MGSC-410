```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
subscriberDF = pd.read_csv('../data/subscriber_USD.csv')

subscriberDF.info()
```

```python
subscriberDF.head()
```

```python
subscriberDF.dtypes
```

```python
subscriberDF = subscriberDF.astype({
    'Language': 'category',
    'Subscription Type': 'string',          # convert to bool, limited or not ?
    'Subscription Event Type': 'string',    # convert to bool, initial_purchase or not ?
    'Purchase Store': 'string',             # convert to bool, app_store or not ?
    'Currency': 'string',
    'Subscription Start Date': 'datetime64[ns]', 
    'Subscription Expiration': 'datetime64[ns]', 
    'Demo User': 'bool',
    'Free Trial User': 'bool', 
    'Free Trial Start Date': 'datetime64[ns]', 
    'Free Trial Expiration': 'datetime64[ns]',
    # 'Auto Renew': 'string', # Off, On - convert
    'Country': 'category', 
    'User Type': 'string',                  # convert to bool, consumer or not ?
    'Lead Platform': 'category',              
    'Email Subscriber': 'bool', 
    'Push Notifications': 'bool'
})

subscriberDF['Auto Renew'] = subscriberDF['Auto Renew'].replace({'Off': False, 'On': True})
subscriberDF['Auto Renew'] = subscriberDF['Auto Renew'].astype('bool')
```
